```python
with open('corona_mv.pickle', 'rb') as handle:
    lagus_data = pickle.load(handle)
```

```python
t = np.arange(datetime(2021,1,1),
              datetime(2021,3,16),
              timedelta(days=1)).astype(datetime)
```

```python
lagus_hro = pd.DataFrame(lagus_data["HRO"], index = t)

# Downloaded RKI data
# for reproducing use (~roughly 200mb)
# url = "https://opendata.arcgis.com/datasets/dd4580c810204019a7b8eb3e0b329dd6_0.csv"
# RKIData = pd.read_csv(url, index_col = 8)

RKIData = pd.read_csv('dd4580c810204019a7b8eb3e0b329dd6_0.csv', index_col = 8)
RKIData_hro = RKIData[RKIData.Landkreis=='SK Rostock']
RKIData_hro.index = pd.to_datetime(RKIData_hro.index)
RKIData_hro_day = RKIData_hro.groupby(pd.Grouper(freq='D')).AnzahlFall.sum()
```

```python
f = plt.figure()
plt.title('7 Tage Inzidenz Rostock')
plt.ylabel('7 Tage Inzidenz / 100.000 Einwohner')
plt.xlabel('Datum')
plt.grid()
plt.plot(RKIData_hro_day.index,
         RKIData_hro_day.rolling(7).sum()/2.08,
         '-o', label='RKI')
plt.plot(lagus_hro.index, lagus_hro.values, '-o', label='LaGuS')
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.legend()
f.autofmt_xdate()
plt.xlim([date(2021, 1, 1), date(2021, 3, 15)])
plt.savefig("rostock.png", dpi = 300)
```
